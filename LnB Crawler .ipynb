# 正式爬蟲

```python
chrome_options = Options()
driver=webdriver.Chrome(executable_path='C:/work/selenium_driver_chrome/chromedriver.exe')
driver.get('https://www.lnb.com.tw/market-place')
driver.minimize_window()
driver.maximize_window()
for i in range(2):
    print(i,'-------------------')  #用來看有沒有在動

    pageSource=driver.page_source
    sp = BeautifulSoup(pageSource,'lxml')
    data1 = sp.find_all('tr', attrs={'class': 'loan' or 'loan backed'})
    c_level=sp.find_all('img', attrs={'class': 'rank'})
    for i in range(10):
            
        print('no',i)               #用來看有沒有在動
        q=data1[i].text
        a=q.split('\n')
        while '' in a :
            a.remove('')
        w=c_level[i].get('alt')
        s=w[1:3]
        a.append(s)
        if len(a) == 10 :
            customerdic={'large enterprise':"normal",'rate':a[0],'period':a[1],'loan_type':a[2],'protection_percent':a[3],'unknow':a[4],'completion1':a[5],'completion2':a[6],'investable balance':a[7],'Recruitment situation1':'still going','Recruitment situation2':'still going','credit rank':a[9]}
            customerdata=customerdata.append(customerdic, ignore_index=True)
        elif len(a) == 11:
            if a[0] == '五千大':
                customerdic={'large enterprise':a[0],'rate':a[1],'period':a[2],'loan_type':a[3],'protection_percent':a[4],'unknow':a[5],'completion1':a[6],'completion2':a[7],'investable balance':a[8],'Recruitment situation1':'still going','Recruitment situation2':'still going','credit rank':a[10]}
                customerdata=customerdata.append(customerdic, ignore_index=True)
            else:
                customerdic={'large enterprise':"normal",'rate':a[0],'period':a[1],'loan_type':a[2],'protection_percent':a[3],'unknow':a[4],'completion1':a[5],'completion2':a[6],'investable balance':a[7],'Recruitment situation1':a[8],'Recruitment situation2':a[9],'credit rank':a[10]}
                customerdata=customerdata.append(customerdic, ignore_index=True)            
        elif len(a) == 12 :
            customerdic={'large enterprise':a[0],'rate':a[1],'period':a[2],'loan_type':a[3],'protection_percent':a[4],'unknow':a[5],'completion1':a[6],'completion2':a[7],'investable balance':a[8],'Recruitment situation1':a[9],'Recruitment situation2':a[10],'credit rank':a[11]}
            customerdata=customerdata.append(customerdic, ignore_index=True)
    time.sleep(10)
    driver.find_element_by_class_name('btn-next-page').click()
```
